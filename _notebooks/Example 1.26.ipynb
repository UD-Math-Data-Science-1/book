```python
import pandas as pd

loans = pd.read_csv("../_datasets/loans.csv")
loans = loans[["loan_amnt", "funded_amnt", "int_rate", "title"]]
loans.head(8)
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(loans):
    # Derive column 'derivedCol' from column: 'int_rate'
    # Transform based on the following examples:
    #    int_rate    Output
    # 1: "10.65%" => "10.65"
    loans.insert(3, "interest rate", loans["int_rate"].str.replace("%", ""))
    # Change column type to float64 for column: 'interest rate'
    loans = loans.astype({'interest rate': 'float64'})
    return loans

loans_clean = clean_data(loans.copy())
loans_clean.head()
```
