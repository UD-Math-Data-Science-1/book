```python
import pandas as pd
nba = pd.read_csv("../_datasets/NBA player stats 2023-24.csv")
nba.head()
```

```python
nba = nba[["NAME", "POS"]]
nba.head(10)
```

```python
dum1 = pd.get_dummies(nba, columns=["POS"])
dum1.head(10)
```

```python
dum2 = nba["POS"].str.get_dummies("-")  # split on hyphen
dum2.head(10)
```

```python
nba = pd.concat([nba, dum2], axis="columns")
nba.head(10)
```
