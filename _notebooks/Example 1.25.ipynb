```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(penguins):
    # Replace missing values with the median of each column in: 'bill_length_mm', 'bill_depth_mm' and 2 other columns
    penguins = penguins.fillna({'bill_length_mm': penguins['bill_length_mm'].median(), 'bill_depth_mm': penguins['bill_depth_mm'].median(), 'flipper_length_mm': penguins['flipper_length_mm'].median(), 'body_mass_g': penguins['body_mass_g'].median()})
    # Replace missing values with the most common value of each column in: 'sex'
    # penguins = penguins.fillna({'sex': penguins['sex'].mode()[0]})
    return penguins

penguins_clean = clean_data(penguins.copy())
penguins_clean.head()
```
