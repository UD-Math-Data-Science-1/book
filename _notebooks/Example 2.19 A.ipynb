```python
import seaborn as sns
import pandas as pd
```

```python
cars = sns.load_dataset("mpg")
cars
```

```python
cuts = pd.cut(cars["horsepower"], range(40, 220, 20))
by_hp = cars.groupby(cuts)
```

```python
by_hp["mpg"].count()
```

```python
def keeper(x):
    return len(x) > 29
hp_30 = by_hp.filter(keeper)
hp_30
```

```python
hp_30["horsepower"].describe()
```
