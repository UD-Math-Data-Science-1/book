```python
import seaborn as sns
import pandas as pd
cars = sns.load_dataset("mpg")
cars
```

```python
sns.relplot(data=cars, x="model_year", y="mpg");
```

```python
sns.relplot(data=cars, 
    x="model_year", y="mpg",
    hue="origin", size="weight",
    );
```

```python
sns.relplot(data=cars, 
    x="model_year", y="mpg",
    kind="line", errorbar="sd"
    );
```

```python
columns = [ "mpg", "horsepower", "displacement", "origin" ]

sns.pairplot(data=cars[columns], hue="origin", height=1.5);
```
